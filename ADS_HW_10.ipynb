```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 2000          # Total population
rho = 0.6         # Recovery probability
sigma = 0.01      # Immunity loss probability
tau = 0.25        # Transmission probability per interaction
c = 0.05          # Probability of any two people interacting

# Initial conditions
S, I, R = [1998], [2], [0]
days = 365

# Simulation loop
for day in range(days):
    S_current = S[-1]
    I_current = I[-1]
    R_current = R[-1]

    # Track new states
    new_S, new_I, new_R = S_current, I_current, R_current

    # People who recover
    recoveries = np.random.rand(I_current) < rho
    recovered_today = np.sum(recoveries)
    new_I -= recovered_today
    new_R += recovered_today

    # People who lose immunity
    immunity_loss = np.random.rand(R_current) < sigma
    susceptible_again = np.sum(immunity_loss)
    new_R -= susceptible_again
    new_S += susceptible_again

    # Potential new infections
    new_infections = 0
    for _ in range(S_current):
        # Number of interactions with infected people
        num_interactions = np.random.binomial(I_current, c)
        # If any result in infection
        if 1 - (1 - tau) ** num_interactions < np.random.rand():
            new_infections += 1

    new_S -= new_infections
    new_I += new_infections

    S.append(new_S)
    I.append(new_I)
    R.append(new_R)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(S, label='Susceptible')
plt.plot(I, label='Infectious')
plt.plot(R, label='Recovered')
plt.xlabel('Day')
plt.ylabel('Number of People')
plt.title('SIR Simulation ')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
